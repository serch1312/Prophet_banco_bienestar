```python
from neuralprophet import NeuralProphet
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score, mean_absolute_error
import numpy as np

# ============================================================
# 1. PREPARAR DATOS
# ============================================================
# Definir periodo de análisis y cajero a modelar
fecha_inicio = '2024-07-01'  # Inicio post-periodo electoral
cajero_test = 'JF000001'

# Filtrar datos del cajero y periodo específico
df_cajero = df[(df['fecha'] >= fecha_inicio) & (df['cajero'] == cajero_test)][['fecha', 'retiro']].copy()
df_cajero.columns = ['ds', 'y']  # NeuralProphet también usa 'ds' e 'y'
df_cajero = df_cajero.sort_values('ds').reset_index(drop=True)

print(f"Cajero {cajero_test}: {len(df_cajero)} observaciones desde {fecha_inicio}")
print(f"Rango: {df_cajero['ds'].min()} → {df_cajero['ds'].max()}")

# ============================================================
# 2. CONFIGURAR HOLIDAYS (FECHAS DE DISPERSIÓN)
# ============================================================
# Elegir nivel de granularidad de holidays:
# - 'granular': 54 eventos (adultos_mayores_A, adultos_mayores_B, etc.)
# - 'agregado': 3 eventos (adultos_mayores, discapacidad, madres_trabajadoras)
holidays_type = 'agregado'
holidays_usar = holidays_df_granular if holidays_type == 'granular' else holidays_df_agregado

print(f"\nHolidays: {holidays_type}")
print(f"  Eventos únicos: {holidays_usar['holiday'].nunique()}")
print(f"  Total registros: {len(holidays_usar)}")

# NeuralProphet requiere que los holidays estén como columnas binarias en el dataframe
df_cajero_eventos = df_cajero.copy()
for holiday_name in holidays_usar['holiday'].unique():
    fechas_holiday = holidays_usar[holidays_usar['holiday'] == holiday_name]['ds'].values
    # Crear columna binaria: 1 si es día de ese holiday, 0 si no
    df_cajero_eventos[holiday_name] = df_cajero_eventos['ds'].isin(fechas_holiday).astype(int)
    # Aplicar ventana de efecto (lower_window=0, upper_window=3)
    for i in range(1, 4):  # 3 días después
        df_cajero_eventos.loc[df_cajero_eventos['ds'].isin(fechas_holiday + pd.Timedelta(days=i)), holiday_name] = 1

print(f"  Columnas de holidays agregadas: {holidays_usar['holiday'].nunique()}")

# ============================================================
# 3. CONFIGURAR MODELO NEURALPROPHET
# ============================================================
print("\n" + "="*80)
print("CONFIGURANDO NEURALPROPHET CON PARÁMETROS TRADUCIDOS DE PROPHET")
print("="*80)

modelo = NeuralProphet(
    # --- TENDENCIA ---
    growth='linear',                      # Tendencia lineal (igual que Prophet)
    n_changepoints=25,                    # Número de puntos de cambio en la tendencia
    changepoints_range=0.9,               # Rango donde se permiten changepoints (90% de los datos)
    trend_reg=0.5,                        # Regularización de tendencia (0.5=moderada)
                                          # Traducido de Prophet changepoint_prior_scale=0.5
    
    # --- COMPONENTE AUTORREGRESIVO (NUEVO vs Prophet) ---
    n_lags=40,                            # Memoria: usa últimos 40 días de retiros como features
                                          # Crítico para capturar ciclo bimestral (~60 días)
    ar_reg=0.01,                          # Regularización baja del AR: permite usar agresivamente el historial
    
    # --- ESTACIONALIDADES ---
    yearly_seasonality=12,                # Términos de Fourier para ciclo anual (igual que Prophet)
    seasonality_reg=0.1,                  # Regularización de estacionalidad (0.1=moderada)
                                          # Traducido de Prophet seasonality_prior_scale=10 (inverso)
    weekly_seasonality=True,              # Activar estacionalidad semanal
    daily_seasonality=False,              # Desactivar estacionalidad diaria
    seasonality_mode='additive',          # Modo aditivo (efectos se suman)
    
    # --- PREDICCIÓN MULTI-HORIZONTE ---
    n_forecasts=30,                       # Predecir 30 días hacia adelante simultáneamente
                                          # (no iterativo, sino 30 salidas en paralelo)
    
    # --- ENTRENAMIENTO ---
    epochs=200,                           # Iteraciones de entrenamiento (más = mejor ajuste, más lento)
    batch_size=32,                        # Tamaño de batch para entrenamiento
    learning_rate=0.005,                  # Tasa de aprendizaje (0.005=moderada)
    loss_func='Huber',                    # Función de pérdida robusta a outliers
    
    # --- OTROS ---
    normalize='off',                      # No normalizar datos (ya están en escala razonable)
    quantiles=[0.05, 0.95]               # Intervalos de incertidumbre (5% y 95%)
)

print(f"\nParámetros configurados:")
print(f"  trend_reg:           {modelo.config_trend.reg} (moderada)")
print(f"  n_lags:              {modelo.config_ar.n_lags} (memoria de 40 días)")
print(f"  ar_reg:              {modelo.config_ar.reg} ← BAJO para usar historial agresivamente")
print(f"  yearly_seasonality:  {modelo.config_seasonality.yearly_arg}")
print(f"  seasonality_reg:     {modelo.config_seasonality.reg}")
print(f"  n_forecasts:         {modelo.n_forecasts} (predicción multi-horizonte)")
print(f"  epochs:              {modelo.config_train.epochs}")

# ============================================================
# 4. AGREGAR HOLIDAYS (EVENTOS) AL MODELO
# ============================================================
print("\nAgregando holidays como eventos...")
for holiday_name in holidays_usar['holiday'].unique():
    modelo = modelo.add_events(
        holiday_name,
        lower_window=0,       # Efecto comienza el día del evento
        upper_window=3,       # Efecto dura 3 días después
        regularization=0.01   # ← CLAVE: Regularización muy baja (traducido de Prophet 0.001)
                              # Permite coeficientes masivos para capturar picos
    )
print(f"  Eventos agregados: {len(holidays_usar['holiday'].unique())}")
print(f"  event_reg (regularización): 0.01 ← MUY BAJO PARA CAPTURAR PICOS")

# ============================================================
# 5. ENTRENAR MODELO
# ============================================================
print("\n" + "="*80)
print("INICIANDO ENTRENAMIENTO (puede tardar varios minutos)...")
print("="*80)

metrics = modelo.fit(df_cajero_eventos, freq='D')
print("\nEntrenamiento completado.")

# ============================================================
# 6. GENERAR PREDICCIONES
# ============================================================
print("\nGenerando predicciones...")
forecast = modelo.predict(df_cajero_eventos)

# NeuralProphet con n_forecasts=30 genera múltiples columnas: yhat1, yhat2, ..., yhat30
# Usamos yhat1 (predicción a 1 día) para comparar con valores históricos
merged = df_cajero.merge(forecast[['ds', 'yhat1']], on='ds').dropna(subset=['yhat1'])

# ============================================================
# 7. APLICAR CORRECCIÓN: NEGATIVOS A CERO
# ============================================================
print("\nAplicando corrección: predicciones negativas → 0")
negativos_antes = (merged['yhat1'] < 0).sum()
merged['yhat1'] = merged['yhat1'].clip(lower=0)
print(f"  Predicciones negativas corregidas: {negativos_antes}")

# ============================================================
# 8. EVALUAR DESEMPEÑO
# ============================================================
if len(merged) > 0:
    r2 = r2_score(merged['y'], merged['yhat1'])
    mae = mean_absolute_error(merged['y'], merged['yhat1'])
    rmse = np.sqrt(((merged['yhat1'] - merged['y'])**2).mean())
    
    print(f"\n{'='*80}")
    print("RESULTADOS:")
    print(f"  R² (coeficiente de determinación): {r2:.4f}")
    print(f"     → % de varianza explicada por el modelo")
    print(f"  MAE (error absoluto medio):        ${mae:,.0f}")
    print(f"     → Error promedio en pesos")
    print(f"  RMSE (raíz del error cuadrático): ${rmse:,.0f}")
    print(f"     → Penaliza más los errores grandes")
    print("="*80)
    
    # ============================================================
    # 9. GRAFICAR AJUSTE
    # ============================================================
    fig, ax = plt.subplots(figsize=(20, 8))
    
    # Serie real (negro)
    ax.plot(df_cajero['ds'], df_cajero['y'], 
            label='Real', color='black', linewidth=1.5, alpha=0.7)
    
    # Predicción (rojo)
    ax.plot(merged['ds'], merged['yhat1'], 
            label=f'Predicción NeuralProphet (R²={r2:.4f})', color='red', linewidth=1.5, alpha=0.8)
    
    # Marcar fechas de dispersión (líneas verticales azules)
    for fecha in holidays_usar['ds'].unique():
        if fecha >= df_cajero['ds'].min() and fecha <= df_cajero['ds'].max():
            ax.axvline(fecha, color='blue', alpha=0.2, linewidth=0.5)
    
    ax.set_title(f'NeuralProphet (n_lags=40, event_reg=0.01) | Cajero {cajero_test} | Desde {fecha_inicio}', 
                 fontsize=14)
    ax.set_xlabel('Fecha')
    ax.set_ylabel('Retiro ($)')
    ax.legend(loc='upper left')
    ax.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()
    
    # ============================================================
    # 10. GRAFICAR COMPONENTES
    # ============================================================
    fig_comp = modelo.plot_components(forecast)
    plt.tight_layout()
    plt.show()
    
    print("\n" + "="*80)
    print("COMPARACIÓN CON PROPHET:")
    print(f"  Prophet R²:        ~0.6546")
    print(f"  NeuralProphet R²:  {r2:.4f}")
    if r2 > 0.6546:
        print(f"  ✓ MEJORA: +{(r2-0.6546)*100:.2f} puntos porcentuales")
    else:
        print(f"  ✗ NO MEJORÓ: {(r2-0.6546)*100:.2f} puntos porcentuales")
    print("="*80)
else:
    print("\n⚠️ ERROR: No se pudieron generar predicciones comparables")
```
